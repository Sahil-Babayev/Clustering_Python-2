# K-Means Clustering

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(data)
```

```python
y_kmeans
```

```python
kmeans.cluster_centers_
```

```python
data['Classes']=y_kmeans
```

```python
data
```

```python
data[(data['Classes']==0)].head(50)
```

```python
data[(data['Classes']==1)].head(50)
```

```python
data = data.replace({0: 'Iris-versicolor', 1: 'Iris-setosa', 2: 'Iris-virginica'})
```

# Hierarchical Clustering

```python
df
```

```python
new_data = df.set_index('Species')
```

```python
new_data.head()
```

```python
dendogram=sch.dendrogram(sch.linkage(new_data, method='ward')) # Within cluster variance is reduced with ward method
plt.title('Dendogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
hc=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y_hc=hc.fit_predict(new_data)
```

```python
y_hc
```

```python
new_data['Classes_new']=y_hc
```

```python
new_data
```

```python
new_data[(new_data['Classes_new']==0)].head(50)
```
